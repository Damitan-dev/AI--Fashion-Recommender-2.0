## Data Loading & Wrangling with NumPy and Pandas

```python
#Import libraries
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns





#Define the file paths
articles_path = '../data/articles.csv'
customers_path = '../data/customers.csv'

#Load into Dataframes
articles_df = pd.read_csv(articles_path)
customers_df = pd.read_csv(customers_path)



```

#### Inspection of the dataframe or dataset

```python
#Inspect the articles_df dataframe
print("--- First 5 rows of articles_df ---")
display(articles_df.head())


print('\n--- Info for articles_df ---')
articles_df.info()

print("--- Shape of the dataset ---")
articles_df.shape
```

```python
# Inspect customers_df Dataframe

print("--- Statistical summary for customers_df ---")
display(customers_df.describe())

#check for missing values

print("\n--- Missing values in customers_df")

print(customers_df.isnull().sum())
```

#### Finding the average age of customers ignoring any missing age values(NaNs)



```python
#Get the 'age'  column,which might contain NaNs

ages = customers_df["age"].values

#Use NumPy's nanmean to calculate the mean, safely ignoring NaNs 
average_age = np.nanmean(ages)



print(f"\nThe average customer age is: {average_age:.2f}")
```

#### Data Exploratory and Visualization

```python
#Set  a style for seaborn plots
sns.set_style('whitegrid')


# Set the size of the plot
plt.figure(figsize=(12,6))

#Create the count plot
sns.countplot(y=articles_df['product_group_name'], order=articles_df['product_group_name'].value_counts().index)


#Add titles and labels
plt.title('Number of products per Product Group')
plt.xlabel('Count')
plt.ylabel('Product Group')


# Show the plot
plt.show()
```

The plot shows that 'Garment Upper body' is by far the most numerous product group.

```python
plt.figure(figsize=(10,6))

# Create the histogram
sns.histplot(data=customers_df, x='age', bins=30, kde=True)

# Add titles and labels
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Customers')

# Show the plot
plt.show()

```

"The customer base is heavily skewed towards a younger demographic, with a large peak between 20-30 years old."
