## Data Loading & Wrangling with NumPy and Pandas

```python
#Import libraries
import pandas as pd
import numpy as np




#Define the file paths
articles_path = '../data/articles.csv'
customers_path = '../data/customers.csv'

#Load into Dataframes
articles_df = pd.read_csv(articles_path)
customers_df = pd.read_csv(customers_path)



```

#### Inspection of the dataframe or dataset

```python
#Inspect the articles_df dataframe
print("--- First 5 rows of articles_df ---")
display(articles_df.head())


print('\n--- Info for articles_df ---')
articles_df.info()

print("--- Shape of the dataset ---")
articles_df.shape
```

```python
# Inspect customers_df Dataframe

print("--- Statistical summary for customers_df ---")
display(customers_df.describe())

#ccheck for missing values

print("\n--- Missing values in customers_df")

print(customers_df.isnull().sum())
```

#### Finding the average age of customers ignoring any missing age values(NaNs)



```python
#Get the 'age'  column,which might contain NaNs

ages = customers_df["age"].values

#Use NumPy's nanmean to calculate the mean, safely ignoring NaNs 
average_age = np.nanmean(ages)



print(f"\nThe average customer age is: {average_age:.2f}")
```
